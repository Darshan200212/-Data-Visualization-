# Forbes Richest Athletes 

**Data Dictionary**
1. Name: Name of the athlete.
2. Nationality: Name of the country the athlete belongs to.
3. Current Rank: Rank in the corresponding Year.
4. Previous Year Rank: Rank in in the preceding year.
5. Sport: Name of the sports that athlete played that year.
6. Year: Current year of ranking.
7. earnings ( million) : Earnings of the athlete in millions in the corresponding year. 

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_palette("deep")
from warnings import filterwarnings
filterwarnings("ignore")
```

```python
df = pd.read_csv("Forbes+Richest+Athletes+1990-2019.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.shape
```

# Atheletes from which country have been listed most number of times in the Forbes Richest Athletes from 1990 to 2019?
[Hint: use countplot]

```python
df["Nationality"].max()
```

```python
plt.xticks(rotation=90)
sns.countplot(x="Nationality",data=df,palette="Accent")
```

# Make a plot visualizing year wise top earning (rank-1) athletes. In which year the rank 1 athlete earned the most? Which was the athlete who earned the most and what sport did he play? How much did the athlete earn?
[hint : use barplot]

```python
plt.figure(figsize=(10,6))
a=df[df["Current Rank"]==1]
a.head()
plt.xticks(rotation=90)
sns.barplot(x="Year",y="earnings ($ million)",hue="Sport",data=a,palette="rocket")
```

# Which athlete has been most frequently ranked number 1 from 1990 to 2019? 
[Hint : use piechart]

```python
a=df[df["Current Rank"]==1]
c=a["Name"].value_counts()
plt.pie(c,autopct="%1.1f%%",labels=c.index,explode=[0.1,0,0,0,0,0,0,0])
```

# Players from which 2 sports have been listed least number of times in the Forbes RIchest Athletes List?
[Hint : Use swarmplot]

```python
plt.figure(figsize=(10,5))
plt.xticks(rotation=90)
sns.swarmplot(x="Sport",y="earnings ($ million)",data=df,palette="Accent")
```

#  Discuss the relationship between Current rank and earnings with the help of a suitable plot.
[hint : use regplot]

```python
df1=df[["Current Rank","earnings ($ million)"]]
df1.head()
```

```python
coor=df1.corr(method="pearson")
coor
```

```python
sns.heatmap(coor,annot=True)
```

#  Which three sports have the highest spread in terms of earnings (highest IQR)?
[Hint : use boxplot]

```python
plt.figure(figsize=(10,7))
plt.xticks(rotation=90)
sns.boxplot(x="Sport",y="earnings ($ million)",data=df)
```

infer you that the Boxing is the higest spread in termas of earnings
